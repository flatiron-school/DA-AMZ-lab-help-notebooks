## Exporting Functions to Separate File

The final notebook does not need to include all of the code you wrote for the project. To make the final notebook more presentable we can save our larger chunks of code in a separate Python file and import it into our notebook.

Inside this project repo I created a Python file called `functions.py` inside of a `src` folder. The code below allows you to import those functions into this notebook to be used without having the code itself.

Not only does this make the notebook cleaner but it also make for more efficient coding. In large projects where you might be working across multiple notebooks you can save your functions in a file and import them into each of your notebooks rather than copy and pasting them into each. 

The following code allows you to make changes to the external Python file and have those changes automatically detected in the notebook. It is not completely necessary but if you don't have it, you will have to restart the kernel every time you make changes to the .py file. I recommend having it. 
```
%load_ext autoreload
%autoreload 2
import os
import sys
module_path = os.path.abspath(os.path.join(os.pardir))
if module_path not in sys.path:
    sys.path.append(module_path)
```

`import src.functions as funcs` imports all the functions from the external file into this notebook. You can think of `funcs` as a variable containing all of the functions defined in the file.

We haven't talked about docstrings yet, but it is best practice (and will be expected in the industry) to add documentation to all of your functions. This article does a great job of describing [Documenting Python Code](https://realpython.com/documenting-python-code/).

````python
import pandas as pd

# Use functions from .py file
%load_ext autoreload
%autoreload 2
import os
import sys
module_path = os.path.abspath(os.path.join(os.pardir))
if module_path not in sys.path:
    sys.path.append(module_path)

import src.functions as funcs
````

````python
df = pd.read_csv('../data/life-expectancy.csv')
df.head()
````

````python
df.columns = [col.strip().replace(' ', '_').lower() for col in df.columns]
df.head()
````

The py file has a function called `plot_hist`. To use this function, I will reference the `funcs` variable followed by a period and the function name.

`funcs.function_name()`

Then I can pass in whatever arguments are required by the function.

````python
funcs.plot_hist(df, 'population', 'Distribution of Population', 'Population')
````
